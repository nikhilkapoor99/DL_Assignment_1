```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, utils
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from scipy.io import loadmat
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Setting a random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

## Loading the Dataset

```python
import urllib.request

# URLs for the train and test dataset .mat files
train_url = "http://ufldl.stanford.edu/housenumbers/train_32x32.mat"
test_url = "http://ufldl.stanford.edu/housenumbers/test_32x32.mat"

# Download the datasets
urllib.request.urlretrieve(train_url, 'train_32x32.mat')
urllib.request.urlretrieve(test_url, 'test_32x32.mat')
```

```python
def load_svhn_data(path):
    data = loadmat(path)
    return data['X'], data['y']

train_images, train_labels = load_svhn_data('train_32x32.mat')
test_images, test_labels = load_svhn_data('test_32x32.mat')

# Transpose the image arrays
train_images = np.transpose(train_images, (3, 0, 1, 2))
test_images = np.transpose(test_images, (3, 0, 1, 2))

# Replace label 10 with label 0
train_labels[train_labels == 10] = 0
test_labels[test_labels == 10] = 0

# Display the shape of the datasets
print("Training images shape:", train_images.shape)
print("Training labels shape:", train_labels.shape)
print("Test images shape:", test_images.shape)
print("Test labels shape:", test_labels.shape)
```

```python
# Checking the distribution of labels in the training set
label_counts_train = np.unique(train_labels, return_counts=True)
label_counts_test = np.unique(test_labels, return_counts=True)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x=label_counts_train[0], y=label_counts_train[1])
plt.title("Distribution of Labels in Training Set")
plt.xlabel("Label")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
sns.barplot(x=label_counts_test[0], y=label_counts_test[1])
plt.title("Distribution of Labels in Test Set")
plt.xlabel("Label")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

## Preprocessing of the Dataset

```python
# Checking for missing values in the images and labels
missing_train_images = np.isnan(train_images).any()
missing_test_images = np.isnan(test_images).any()

missing_train_labels = np.isnan(train_labels).any()
missing_test_labels = np.isnan(test_labels).any()

print(f"Missing values in training images: {missing_train_images}")
print(f"Missing values in test images: {missing_test_images}")
print(f"Missing values in training labels: {missing_train_labels}")
print(f"Missing values in test labels: {missing_test_labels}")
```

```python
# Normalize the images to [0, 1] range
train_images = train_images.astype("float32") / 255.0
test_images = test_images.astype("float32") / 255.0

# Check the shape of the labels before one-hot encoding
print("Before one-hot encoding:", train_labels.shape)

# Convert labels to one-hot encoded format
train_labels = utils.to_categorical(train_labels, 10)
test_labels = utils.to_categorical(test_labels, 10)

# Check the shape of the labels after one-hot encoding
print("After one-hot encoding:", train_labels.shape)
```

```python
from sklearn.model_selection import train_test_split

# Splitting the training data into training and validation sets
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)

# Display the shape of the resulting datasets
print("Training images shape:", train_images.shape)
print("Validation images shape:", val_images.shape)
print("Training labels shape:", train_labels.shape)
print("Validation labels shape:", val_labels.shape)
```

## Constructing the CNN Model

First, We've constructed a baseline Convolutional Neural Network (CNN) for the SVHN dataset. The model consists of three convolutional layers, each followed by a max-pooling layer. These layers are designed to capture spatial hierarchies and reduce dimensionality. Post these layers, the model flattens the output and passes it through a dense layer with 64 neurons. The final layer has 10 neurons corresponding to the 10 classes in the SVHN dataset, using a softmax activation to produce class probabilities. This simple architecture serves as our starting point to gauge performance before exploring more complex models.

```python
# Define the CNN model
# Baseline Model
model_bl = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Print the model summary
model_bl.summary()
```

```python
# Compile the model
model_bl.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model using the validation set
history_bl = model_bl.fit(train_images, train_labels, epochs=20,
                    validation_data=(val_images, val_labels))
```

```python
# Evaluate the model on the test set
test_loss_bl, test_acc_bl = model_bl.evaluate(test_images, test_labels, verbose=2)
print("\nTest accuracy:", test_acc_bl)
print("\nTest loss:", test_loss_bl)
```

```python
# Plotting training & validation accuracy values
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_bl.history['accuracy'])
plt.plot(history_bl.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plotting training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history_bl.history['loss'])
plt.plot(history_bl.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Predicting the labels for the test set
y_pred_bl = model_bl.predict(test_images)
y_pred_classes_bl = np.argmax(y_pred_bl, axis=1)
y_true_bl = np.argmax(test_labels, axis=1)

# Generating the confusion matrix
confusion_mtx_bl = confusion_matrix(y_true_bl, y_pred_classes_bl)

# Plotting the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_mtx_bl, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Printing the classification report
print(classification_report(y_true_bl, y_pred_classes_bl))
```

```python
import optuna
from optuna.integration import TFKerasPruningCallback
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Dropout, Flatten, Dense

def objective(trial):
    # Define hyperparameters to be tuned
    lr = trial.suggest_float('lr', 1e-5, 1e-2, log=True)
    num_filters_1 = trial.suggest_int('num_filters_1', 16, 128, log=True)
    num_filters_2 = trial.suggest_int('num_filters_2', 16, 128, log=True)
    dropout_rate_1 = trial.suggest_float('dropout_rate_1', 0.1, 0.5)
    dropout_rate_2 = trial.suggest_float('dropout_rate_2', 0.1, 0.5)
    batch_size = trial.suggest_categorical('batch_size', [32, 128])
    dense_units = trial.suggest_int('dense_units', 64, 512, log=True)

    # Build the model using the suggested hyperparameters
    model_bo = Sequential([
        Conv2D(num_filters_1, (3, 3), activation='relu', input_shape=(32, 32, 3)),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(dropout_rate_1),

        Conv2D(num_filters_2, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D((2, 2)),
        Dropout(dropout_rate_2),

        Flatten(),
        Dense(dense_units, activation='relu'),
        BatchNormalization(),
        Dropout(0.4),

        Dense(10, activation='softmax')
    ])

    model_bo.compile(optimizer=Adam(learning_rate=lr),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # Train the model using the suggested batch size and add the pruning callback
    history_bo = model_bo.fit(
        train_images, train_labels,
        epochs=20,
        batch_size=batch_size,
        validation_data=(val_images, val_labels),
        verbose=1,
        callbacks=[TFKerasPruningCallback(trial, 'val_accuracy')]
    )

    # Return the validation accuracy of the last epoch
    return history_bo.history['val_accuracy'][-1]

# Create a study with the MedianPruner
pruner = optuna.pruners.MedianPruner()
study = optuna.create_study(direction='maximize', pruner=pruner)

# Run the optimization with the specified number of trials
study.optimize(objective, n_trials=20)

# Get the best hyperparameters
best_params = study.best_params
best_accuracy = study.best_value
print(f"Best hyperparameters: {best_params}")
print(f"Best validation accuracy: {best_accuracy}")
```

```python
# Define the model with the best hyperparameters
best_params = {'lr': 0.0004974088484584535, 'num_filters_1': 109, 'num_filters_2': 28, 'dropout_rate_1': 0.12464363952996807, 'dropout_rate_2': 0.31255579482360896, 'batch_size': 32, 'dense_units': 214}

model_best = Sequential([
    Conv2D(best_params['num_filters_1'], (3, 3), activation='relu', input_shape=(32, 32, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(best_params['dropout_rate_1']),

    Conv2D(best_params['num_filters_2'], (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(best_params['dropout_rate_2']),

    Flatten(),
    Dense(best_params['dense_units'], activation='relu'),
    BatchNormalization(),
    Dropout(0.4),

    Dense(10, activation='softmax')
])

model_best.compile(optimizer=Adam(learning_rate=best_params['lr']),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model using the best batch size
history_best = model_best.fit(
    train_images, train_labels,
    epochs=20,
    batch_size=best_params['batch_size'],
    validation_data=(val_images, val_labels),
    verbose=1
)
```

```python
# Accuracy of the Best Model
test_loss_best, test_accuracy_best = model_best.evaluate(test_images, test_labels)
print(f"Test Accuracy: {test_accuracy_best}")
print(f"Test Loss: {test_loss_best}")
```

## Appendex - A statistical test

This test was permormed between best advanced architecture and the retrained custom model

```python
from statsmodels.stats.contingency_tables import mcnemar

# Predictions from both models
y_pred_mobilenet_classes = np.argmax(model_mobilenet.predict(test_images), axis=1)
y_pred_best_classes = np.argmax(model_best.predict(test_images), axis=1)
y_true_classes = np.argmax(test_labels, axis=1)

# Create contingency table
table = np.zeros((2, 2))

for i in range(len(y_true_classes)):
    true = y_true_classes[i]
    pred_mobilenet = y_pred_mobilenet_classes[i]
    pred_best = y_pred_best_classes[i]

    if pred_mobilenet == true and pred_best == true:
        table[0, 0] += 1
    elif pred_mobilenet == true and pred_best != true:
        table[0, 1] += 1
    elif pred_mobilenet != true and pred_best == true:
        table[1, 0] += 1
    else:
        table[1, 1] += 1

# Apply McNemar's test
result = mcnemar(table, exact=True)

# Print the result
print(f"Statistic: {result.statistic}")
print(f"p-value: {result.pvalue}")

# Determine significance
alpha = 0.05
if result.pvalue > alpha:
    print("There is no significant difference between the two models (fail to reject H0)")
else:
    print("There is a significant difference between the two models (reject H0)")
```
